# Analyzing Star Wars Survey Data

Before Star Wars: The Force Awakens came out, FiveThirtyEight conducted a survey on Star Wars fans using the online tool SurveyMonkey. The team wondered questions such as: does the rest of America realize that "The Empire Strikes Back" is clearly the best of the bunch? 

The [survey](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey) consists of 835 total responses which we will be cleaning and exploring in this project.

## Read in the data

```python
import pandas as pd
import numpy as np
star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')
```

## Data exploration

```python
star_wars.head(10)
```

```python
star_wars.columns
```

## Removing rows where RespondentID is NaN

```python
star_wars = star_wars[star_wars['RespondentID'].notnull()]
```

```python
star_wars.head()
```

## Cleaning and Mapping Yes/No Columns

```python
yes_no = {'Yes': True, 'No': False}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] =  star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] =  star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
```

```python
star_wars.head()
```

## Cleaning and Mapping Checkbox Columns

The next six columns represent a single checkbox question. The respondent checked off a box for each Star Wars film they have seen. We will clean and rename these columns to make them more intuitive and easier to analyze.

```python
movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

```python
star_wars = star_wars.rename(columns={'Which of the following Star Wars films have you seen? Please select all that apply.': 'seen_1',
                                     'Unnamed: 4': 'seen_2',
                                     'Unnamed: 5': 'seen_3', 
                                     'Unnamed: 6': 'seen_4',
                                     'Unnamed: 7': 'seen_5',
                                     'Unnamed: 8': 'seen_6'})
```

```python
star_wars.head()
```

## Cleaning the Ranking Columns

The next six columns are the resondent's ranking for each Star Wars movie in order of least to favourite. For these columns we will convert each column to a numeric type and then rename the columns to make them easier to understand.

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars = star_wars.rename(columns={'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'ranking_1',
                                     'Unnamed: 10': 'ranking_2',
                                     'Unnamed: 11': 'ranking_3',
                                     'Unnamed: 12': 'ranking_4',
                                     'Unnamed: 13': 'ranking_5',
                                     'Unnamed: 14': 'ranking_6'})
```

## Finding the Highest-Ranked Movie

```python
ranking_means = star_wars[star_wars.columns[9:15]].mean()
ranking_means
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots()
ranking_bar = ax.bar(range(6), ranking_means)
ax.set_title('Rankings For Each Star Wars Movie')
plt.show()
```

After creating a bar chart to visualize the average ranking for each film, it is evident that the original movies are ranked higher than the prequel triology.

## Finding the Most Viewed Movie

```python
views = star_wars[star_wars.columns[3:9]].sum()
views
```

```python
fig, ax = plt.subplots()
ax.set_title('Views For Each Star Wars Movie')
views_bar = ax.bar(range(6), views)
```

As shown on the above bar chart, the original movies have more views than the prequels. This confirms that the original movies are more popular than the prequels.

## Exploring the Data by Binary Segments

```python
males = star_wars[star_wars['Gender'] == 'Male']
females = star_wars[star_wars['Gender'] == 'Female']
```

```python
ranking_means_males = males[males.columns[9:15]].mean()
ranking_means_females = females[females.columns[9:15]].mean()

fig, ax = plt.subplots(2, figsize=(7,10))
ranking_bar_males = ax[0].bar(range(6), ranking_means_males)
ax[0].set_title('Male Rankings')
ranking_bar_females = ax[1].bar(range(6), ranking_means_females)
ax[1].set_title('Female Rankings')
plt.show()
```

The rankings for each Star Wars film seem to be very similar in both Male and Female segments of the dataset.

```python
views_males = males[males.columns[3:9]].sum()
views_females = females[females.columns[3:9]].sum()

fig, ax = plt.subplots(2, figsize=(7,10))
views_bar_males = ax[0].bar(range(6), views_males)
ax[0].set_title('Male Views')
views_bar_females = ax[1].bar(range(6), views_females)
ax[1].set_title('Female Views')
plt.show()
```

Males have slightly higher views for each Star Wars Film. We can also identify that more Males watched the prequels than females.
